```python
# Identificar colunas de tempo em df_tempo_mac
cols_tempo_mac = [col for col in df_tempo_mac.columns if 'Valor' in col and col != 'Valor_numero']
if not cols_tempo_mac:  # Se n√£o encontrar, usar apenas Valor_numero
    cols_tempo_mac = ['Valor_numero']

df_tempo_mac['Valor'] = df_tempo_mac[cols_tempo_mac].mean(axis=1)
df_tempo_mac['Valor_std'] = df_tempo_mac[cols_tempo_mac].std(axis=1)
df_tempo_mac['Valor_delta'] = (
    df_tempo_mac[cols_tempo_mac].max(axis=1) -
    df_tempo_mac[cols_tempo_mac].min(axis=1)
)

df_resumo['Temp_media'] = df_resumo[cols_temp_A].mean(axis=1)
df_resumo['Temp_std'] = df_resumo[cols_temp_A].std(axis=1)
df_resumo['Temp_delta'] = (
    df_resumo[cols_temp_A].max(axis=1) -
    df_resumo[cols_temp_A].min(axis=1)
)
```

```python
if 'MesAno_dt1' not in df_tempo_mac.columns:
    df_tempo_mac['MesAno_dt1'] = pd.to_datetime(
        df_tempo_mac['M√™s/Ano'], 
        format='%b/%y', 
        errors='coerce'
    )

# Filtrar per√≠odo 2024-2026
df_tempo_2426 = df_tempo_mac[
    (df_tempo_mac['MesAno_dt1'] >= '2024-01-01') &
    (df_tempo_mac['MesAno_dt1'] <= '2026-12-31')
].copy()

print(f"üìä Total de registros filtrados (2024-2026): {len(df_tempo_2426)}")
print(f"   Per√≠odo: {df_tempo_2426['MesAno_dt1'].min().strftime('%b/%Y')} at√© {df_tempo_2426['MesAno_dt1'].max().strftime('%b/%Y')}")

# Verificar meses √∫nicos
meses_unicos = df_tempo_2426[['M√™s/Ano', 'MesAno_dt1']].drop_duplicates().sort_values('MesAno_dt1')
print(f"\n   Total de meses: {len(meses_unicos)}")
```

```python
# 1. Preparar dados de temperatura do df_resumo (agrupado por per√≠odo)
# Primeiro, criar coluna de per√≠odo no df_resumo se n√£o existir
if 'MesAno_dt1' not in df_resumo.columns:
    df_resumo['MesAno_dt1'] = pd.to_datetime(
        df_resumo[col_mes], 
        format='%b/%y', 
        errors='coerce'
    )

# Filtrar per√≠odo 2024-2026 para temperatura
df_temp_2426 = df_resumo[
    (df_resumo['MesAno_dt1'] >= '2024-01-01') &
    (df_resumo['MesAno_dt1'] <= '2026-12-31')
].copy()

# 2. Agrupar temperatura por m√™s
temp_mensal = df_temp_2426.groupby('MesAno_dt1').agg({
    'Temp_media': 'mean',
    'Temp_std': 'mean'
}).sort_index()

# 3. Agrupar tempo por m√™s (j√° filtrado em df_tempo_2426)
tempo_mensal = df_tempo_2426.groupby('MesAno_dt1').agg({
    'Valor_numero': 'mean'
}).sort_index()

# 4. Criar um DataFrame alinhado com ambos os dados
df_alinhado = pd.DataFrame({
    'Temperatura': temp_mensal['Temp_media'],
    'Temp_std': temp_mensal['Temp_std'],
    'Tempo': tempo_mensal['Valor_numero']
})

# Remover valores nulos
df_alinhado = df_alinhado.dropna()

print(f"üìä Dados alinhados por per√≠odo:")
print(f"   Total de meses: {len(df_alinhado)}")
print(f"   Per√≠odo: {df_alinhado.index.min().strftime('%b/%Y')} at√© {df_alinhado.index.max().strftime('%b/%Y')}")
print(f"\n   Primeiros registros:")
print(df_alinhado.head())

# 5. Criar gr√°fico com dois eixos Y (temperatura e tempo)
fig, ax1 = plt.subplots(figsize=(14, 6))

# Eixo 1: Temperatura
color1 = '#AB362E'
ax1.set_xlabel('Per√≠odo', fontsize=12)
ax1.set_ylabel('Temperatura (¬∞C)', color=color1, fontsize=12)
ax1.plot(df_alinhado.index, df_alinhado['Temperatura'], 
         marker='o', linewidth=2, color=color1, markersize=8, label='Temperatura')
ax1.fill_between(df_alinhado.index,
                  df_alinhado['Temperatura'] - df_alinhado['Temp_std'],
                  df_alinhado['Temperatura'] + df_alinhado['Temp_std'],
                  alpha=0.2, color=color1)
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, alpha=0.3)

# Eixo 2: Tempo
ax2 = ax1.twinx()
color2 = '#2E5AAB'
ax2.set_ylabel('Tempo de Macera√ß√£o (min)', color=color2, fontsize=12)
ax2.plot(df_alinhado.index, df_alinhado['Tempo'], 
         marker='s', linewidth=2, color=color2, markersize=8, 
         linestyle='--', label='Tempo')
ax2.tick_params(axis='y', labelcolor=color2)

# T√≠tulo e legendas
plt.title('Temperatura e Tempo de Macera√ß√£o - Alinhados por Per√≠odo', 
          fontsize=14, fontweight='bold', pad=20)
fig.tight_layout()
plt.show()

# Mostrar correla√ß√£o
correlacao = df_alinhado['Temperatura'].corr(df_alinhado['Tempo'])
print(f"\nüìà Correla√ß√£o entre Temperatura e Tempo: {correlacao:.3f}")
```

```python
# GR√ÅFICO DE DISPERS√ÉO COM LINHA DE TEND√äNCIA
from scipy import stats

fig, ax = plt.subplots(figsize=(12, 7))

# Scatter plot
colors = plt.cm.RdYlBu_r(np.linspace(0.3, 0.7, len(df_alinhado)))
scatter = ax.scatter(df_alinhado['Temperatura'], df_alinhado['Tempo'], 
                     s=200, c=colors, alpha=0.6, edgecolors='black', linewidth=2)

# Adicionar labels nos pontos
for idx, row in df_alinhado.iterrows():
    ax.annotate(idx.strftime('%b/%y'), 
                (row['Temperatura'], row['Tempo']),
                xytext=(8, 8), textcoords='offset points',
                fontsize=9, alpha=0.8, fontweight='bold')

# Linha de tend√™ncia (regress√£o linear)
slope, intercept, r_value, p_value, std_err = stats.linregress(
    df_alinhado['Temperatura'], df_alinhado['Tempo']
)

# Criar linha de predi√ß√£o
x_line = np.linspace(df_alinhado['Temperatura'].min(), 
                     df_alinhado['Temperatura'].max(), 100)
y_line = slope * x_line + intercept

ax.plot(x_line, y_line, 'r--', linewidth=2.5, alpha=0.8, 
        label=f'Tend√™ncia: y = {slope:.2f}x + {intercept:.2f}')

# Configura√ß√µes do gr√°fico
ax.set_xlabel('Temperatura (¬∞C)', fontsize=13, fontweight='bold')
ax.set_ylabel('Tempo de Macera√ß√£o (min)', fontsize=13, fontweight='bold')
ax.set_title('Rela√ß√£o Temperatura √ó Tempo de Macera√ß√£o\ncom Linha de Tend√™ncia', 
             fontsize=14, fontweight='bold', pad=20)
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(loc='best', fontsize=11, framealpha=0.9)

# Adicionar informa√ß√µes estat√≠sticas
textstr = f'R¬≤ = {r_value**2:.3f}\nCorrela√ß√£o = {r_value:.3f}\np-value = {p_value:.4f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=11,
        verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()

print(f"üìà Equa√ß√£o da reta: Tempo = {slope:.2f} √ó Temperatura + {intercept:.2f}")
print(f"   R¬≤ = {r_value**2:.3f} (explica {r_value**2*100:.1f}% da varia√ß√£o)")
print(f"   p-value = {p_value:.4f} {'‚úì Significativo' if p_value < 0.05 else '‚úó N√£o significativo'}")
```

## üìã Interpreta√ß√£o Operacional

**Contexto:** A temperatura √© controlada por opera√ß√£o humana, n√£o √© uma vari√°vel ambiental.

### Principais Descobertas:

1. **Temperatura como Vari√°vel de Controle:**
   - Os operadores mant√™m a temperatura relativamente est√°vel (49.65¬∞C ¬± 0.66¬∞C)
   - Coeficiente de varia√ß√£o baixo (1.32%) indica controle operacional consistente

2. **Rela√ß√£o Temperatura √ó Tempo:**
   - Correla√ß√£o: -0.684 (negativa moderada)
   - **Interpreta√ß√£o:** Quando a temperatura √© mantida mais alta (~50¬∞C), o tempo de macera√ß√£o √© menor
   - **Quando a temperatura cai** (~48¬∞C), o tempo aumenta significativamente

3. **Implica√ß√µes Pr√°ticas:**
   - **Jan/2026:** Temperatura mais baixa (48.32¬∞C) ‚Üí Tempo mais longo (80 min)
   - **Mar/2025:** Temperatura mais alta (50.20¬∞C) ‚Üí Tempo mais curto (62.8 min)
   - **Economia de tempo:** ~17 minutos de diferen√ßa entre extremos

4. **Recomenda√ß√µes Operacionais:**
   - Manter temperatura pr√≥xima a 50¬∞C pode reduzir o tempo de macera√ß√£o
   - Avaliar se a qualidade do produto √© afetada pela temperatura
   - Considerar o custo-benef√≠cio: tempo vs. energia (aquecimento)
   - Padronizar o controle de temperatura para ter tempos mais previs√≠veis


```python
# AN√ÅLISE DE POTENCIAL DE OTIMIZA√á√ÉO OPERACIONAL
print("=" * 70)
print("üéØ AN√ÅLISE DE OTIMIZA√á√ÉO DO CONTROLE DE TEMPERATURA")
print("=" * 70)

# Calcular tempo estimado para diferentes temperaturas usando a regress√£o
temp_alvo = 50.0  # Temperatura alvo proposta
tempo_estimado = slope * temp_alvo + intercept

print(f"\n1. SITUA√á√ÉO ATUAL:")
print(f"   ‚Ä¢ Temperatura m√©dia: {df_alinhado['Temperatura'].mean():.2f}¬∞C")
print(f"   ‚Ä¢ Tempo m√©dio: {df_alinhado['Tempo'].mean():.2f} min")

print(f"\n2. CEN√ÅRIO OTIMIZADO (Temperatura = {temp_alvo}¬∞C):")
print(f"   ‚Ä¢ Tempo estimado: {tempo_estimado:.2f} min")
print(f"   ‚Ä¢ Redu√ß√£o potencial: {df_alinhado['Tempo'].mean() - tempo_estimado:.2f} min por batelada")

# An√°lise por faixas de temperatura
df_alinhado['Faixa_Temp'] = pd.cut(df_alinhado['Temperatura'], 
                                     bins=3, 
                                     labels=['Baixa (<49.3¬∞C)', 'M√©dia (49.3-49.9¬∞C)', 'Alta (>49.9¬∞C)'])

print(f"\n3. COMPARA√á√ÉO POR FAIXAS DE TEMPERATURA:")
for faixa in df_alinhado['Faixa_Temp'].cat.categories:
    dados_faixa = df_alinhado[df_alinhado['Faixa_Temp'] == faixa]
    if len(dados_faixa) > 0:
        print(f"\n   {faixa}:")
        print(f"      ‚Ä¢ Tempo m√©dio: {dados_faixa['Tempo'].mean():.1f} min")
        print(f"      ‚Ä¢ N¬∫ de per√≠odos: {len(dados_faixa)}")
        print(f"      ‚Ä¢ Meses: {', '.join([d.strftime('%b/%y') for d in dados_faixa.index])}")

# Calcular impacto anual (estimativa)
print(f"\n4. IMPACTO ESTIMADO (considerando opera√ß√£o cont√≠nua):")
reducao_por_batelada = df_alinhado['Tempo'].mean() - tempo_estimado
bateladas_por_dia = 10  # Ajustar conforme a realidade da opera√ß√£o
dias_por_ano = 300  # Dias operacionais
reducao_anual_horas = (reducao_por_batelada * bateladas_por_dia * dias_por_ano) / 60

print(f"   ‚Ä¢ Redu√ß√£o por batelada: {reducao_por_batelada:.1f} min")
print(f"   ‚Ä¢ Assumindo {bateladas_por_dia} bateladas/dia √ó {dias_por_ano} dias/ano")
print(f"   ‚Ä¢ Economia anual estimada: {reducao_anual_horas:.0f} horas de processo")
print(f"   ‚Ä¢ Equivalente a: {reducao_anual_horas/24:.1f} dias de produ√ß√£o")

print(f"\n5. A√á√ÉO RECOMENDADA:")
print(f"   ‚úì Estabelecer temperatura padr√£o de opera√ß√£o: {temp_alvo}¬∞C ¬± 0.5¬∞C")
print(f"   ‚úì Treinar operadores para manter esse setpoint")
print(f"   ‚úì Monitorar qualidade do produto nessa temperatura")
print(f"   ‚úì Avaliar consumo energ√©tico vs. ganho de produtividade")

print("\n" + "=" * 70)
```

```python
---
# üîÑ AN√ÅLISE INVERTIDA: TEMPO ‚Üí TEMPERATURA

**Objetivo:** Avaliar se o **tempo de macera√ß√£o interfere na temperatura**

**Modelo:**
- **Tempo = Vari√°vel Independente (X)** - ajustado/controlado
- **Temperatura = Vari√°vel Dependente (Y)** - resultado observado

**Hip√≥tese:** Quanto maior o tempo de macera√ß√£o, a temperatura pode cair devido √† perda de calor natural do processo.
```

```python
# AN√ÅLISE DE REGRESS√ÉO: TEMPO ‚Üí TEMPERATURA
from scipy import stats

print("="*70)
print("üîÑ AN√ÅLISE: IMPACTO DO TEMPO NA TEMPERATURA")
print("="*70)

# Regress√£o linear: Tempo (X) ‚Üí Temperatura (Y)
slope_inv, intercept_inv, r_value_inv, p_value_inv, std_err_inv = stats.linregress(
    df_alinhado['Tempo'], df_alinhado['Temperatura']
)

print(f"\nüìä MODELO ESTAT√çSTICO:")
print(f"   Equa√ß√£o: Temperatura = {slope_inv:.4f} √ó Tempo + {intercept_inv:.2f}")
print(f"   Coeficiente Angular: {slope_inv:.4f}¬∞C/min")
print(f"   Intercepto: {intercept_inv:.2f}¬∞C")

print(f"\nüìà QUALIDADE DO MODELO:")
print(f"   R¬≤: {r_value_inv**2:.3f} (explica {r_value_inv**2*100:.1f}% da varia√ß√£o)")
print(f"   Correla√ß√£o (R): {r_value_inv:.3f}")
print(f"   p-value: {p_value_inv:.4f} {'‚úì Significativo (p < 0.05)' if p_value_inv < 0.05 else '‚úó N√£o significativo (p ‚â• 0.05)'}")
print(f"   Erro Padr√£o: {std_err_inv:.5f}")

print(f"\nüîç INTERPRETA√á√ÉO:")
if slope_inv < 0:
    print(f"   ‚Ä¢ Rela√ß√£o NEGATIVA: A cada minuto adicional de macera√ß√£o,")
    print(f"     a temperatura diminui em m√©dia {abs(slope_inv):.4f}¬∞C")
    print(f"   ‚Ä¢ Isso sugere PERDA DE CALOR ao longo do processo")
else:
    print(f"   ‚Ä¢ Rela√ß√£o POSITIVA: A cada minuto adicional de macera√ß√£o,")
    print(f"     a temperatura aumenta em m√©dia {slope_inv:.4f}¬∞C")
    print(f"   ‚Ä¢ Isso pode indicar aquecimento durante o processo")

# Analisar casos extremos
tempo_min = df_alinhado['Tempo'].min()
tempo_max = df_alinhado['Tempo'].max()
temp_pred_min = slope_inv * tempo_min + intercept_inv
temp_pred_max = slope_inv * tempo_max + intercept_inv

print(f"\nüìç PREDI√á√ïES DO MODELO:")
print(f"   ‚Ä¢ Tempo m√≠nimo ({tempo_min:.1f} min) ‚Üí Temperatura prevista: {temp_pred_min:.2f}¬∞C")
print(f"   ‚Ä¢ Tempo m√°ximo ({tempo_max:.1f} min) ‚Üí Temperatura prevista: {temp_pred_max:.2f}¬∞C")
print(f"   ‚Ä¢ Varia√ß√£o temperatura: {abs(temp_pred_max - temp_pred_min):.2f}¬∞C")

# Comparar com dados reais
idx_tempo_min = df_alinhado['Tempo'].idxmin()
idx_tempo_max = df_alinhado['Tempo'].idxmax()
temp_real_min = df_alinhado.loc[idx_tempo_min, 'Temperatura']
temp_real_max = df_alinhado.loc[idx_tempo_max, 'Temperatura']

print(f"\nüìä DADOS REAIS:")
print(f"   ‚Ä¢ Menor tempo ({tempo_min:.1f} min em {idx_tempo_min.strftime('%b/%y')}): Temp = {temp_real_min:.2f}¬∞C")
print(f"   ‚Ä¢ Maior tempo ({tempo_max:.1f} min em {idx_tempo_max.strftime('%b/%y')}): Temp = {temp_real_max:.2f}¬∞C")
print(f"   ‚Ä¢ Diferen√ßa real: {abs(temp_real_max - temp_real_min):.2f}¬∞C")

print(f"\nüí° CONCLUS√ÉO:")
if abs(r_value_inv) < 0.3:
    print(f"   ‚ö†Ô∏è Correla√ß√£o FRACA ({abs(r_value_inv):.3f})")
    print(f"   O tempo explica apenas {r_value_inv**2*100:.1f}% da varia√ß√£o da temperatura.")
    print(f"   Outros fatores podem ter maior influ√™ncia na temperatura.")
elif abs(r_value_inv) < 0.7:
    print(f"   üìä Correla√ß√£o MODERADA ({abs(r_value_inv):.3f})")
    print(f"   O tempo explica {r_value_inv**2*100:.1f}% da varia√ß√£o da temperatura.")
    print(f"   H√° rela√ß√£o, mas outros fatores tamb√©m importam.")
else:
    print(f"   ‚úì Correla√ß√£o FORTE ({abs(r_value_inv):.3f})")
    print(f"   O tempo explica {r_value_inv**2*100:.1f}% da varia√ß√£o da temperatura.")
    print(f"   O tempo tem impacto significativo na temperatura.")

print("="*70)
```

```python
# GR√ÅFICO DE DISPERS√ÉO: TEMPO (X) ‚Üí TEMPERATURA (Y)
fig, ax = plt.subplots(figsize=(14, 8))

# Scatter plot
colors = plt.cm.coolwarm(np.linspace(0.2, 0.8, len(df_alinhado)))
scatter = ax.scatter(df_alinhado['Tempo'], df_alinhado['Temperatura'], 
                     s=300, c=colors, alpha=0.7, edgecolors='black', linewidth=2.5,
                     zorder=5)

# Adicionar labels nos pontos
for idx, row in df_alinhado.iterrows():
    ax.annotate(idx.strftime('%b/%y'), 
                (row['Tempo'], row['Temperatura']),
                xytext=(10, 10), textcoords='offset points',
                fontsize=11, alpha=0.9, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.3))

# Linha de regress√£o
tempo_range = np.linspace(df_alinhado['Tempo'].min() - 2, 
                          df_alinhado['Tempo'].max() + 2, 100)
temp_pred = slope_inv * tempo_range + intercept_inv

ax.plot(tempo_range, temp_pred, 'r-', linewidth=3.5, alpha=0.8, 
        label=f'Modelo: Temp = {slope_inv:.4f} √ó Tempo + {intercept_inv:.2f}', zorder=4)

# Intervalo de confian√ßa (simplificado)
residuals = df_alinhado['Temperatura'] - (slope_inv * df_alinhado['Tempo'] + intercept_inv)
std_residuals = residuals.std()
ax.fill_between(tempo_range, temp_pred - 1.96*std_residuals, temp_pred + 1.96*std_residuals,
                alpha=0.2, color='red', label='Intervalo de Confian√ßa (95%)')

# Configura√ß√µes do gr√°fico
ax.set_xlabel('Tempo de Macera√ß√£o (min)', fontsize=14, fontweight='bold')
ax.set_ylabel('Temperatura (¬∞C)', fontsize=14, fontweight='bold')
ax.set_title('An√°lise: Impacto do Tempo de Macera√ß√£o na Temperatura\n(Tempo como Vari√°vel Independente)', 
             fontsize=16, fontweight='bold', pad=20)
ax.grid(True, alpha=0.4, linestyle='--', linewidth=1)
ax.legend(loc='best', fontsize=12, framealpha=0.95, shadow=True)

# Adicionar estat√≠sticas no gr√°fico
textstr = f'R¬≤ = {r_value_inv**2:.3f}\n'
textstr += f'Correla√ß√£o = {r_value_inv:.3f}\n'
textstr += f'p-value = {p_value_inv:.4f}\n'
if slope_inv < 0:
    textstr += f'\n‚ö†Ô∏è Perda de calor:\n{abs(slope_inv):.4f}¬∞C por minuto'
else:
    textstr += f'\nüî• Ganho de calor:\n{slope_inv:.4f}¬∞C por minuto'

props = dict(boxstyle='round,pad=0.8', facecolor='lightblue', alpha=0.9, edgecolor='navy', linewidth=2)
ax.text(0.02, 0.98, textstr, transform=ax.transAxes, fontsize=12,
        verticalalignment='top', bbox=props, family='monospace')

# Destacar casos extremos
ax.scatter([tempo_min], [temp_real_min], s=500, c='green', marker='*', 
           edgecolors='darkgreen', linewidth=3, zorder=10, label='Menor Tempo')
ax.scatter([tempo_max], [temp_real_max], s=500, c='orange', marker='*', 
           edgecolors='darkorange', linewidth=3, zorder=10, label='Maior Tempo')

ax.legend(loc='best', fontsize=11, framealpha=0.95, shadow=True)

plt.tight_layout()
plt.show()

print(f"üìä Gr√°fico gerado com sucesso!")
print(f"   Eixo X: Tempo (vari√°vel independente)")
print(f"   Eixo Y: Temperatura (vari√°vel dependente)")
```
